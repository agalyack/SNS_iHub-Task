```python
from langchain.llms import CTransformers
```

```python
model_file='/content/drive/MyDrive/llama-2-7b-chat.ggmlv3.q8_0.bin'
```

```python
llm_model=CTransformers(model=model_file,model_type='llama',config={"max_new_tokens":500,"temperature":0.01})
```

```python
llm_model
```

```python
prompt = """
use the following information and answer the question based on that /
provide only useful informations /
if you do not know, do not try to make wrong answers /
if you do not know, just respond as 'do not know'/

show only the answer alone /
Example: what is the name in doc?
         output: Raj

Context: {context}
Questions: {question}

"""
```

```python
from langchain import PromptTemplate
from langchain.chains import RetrievalQA
```

```python
def squ_prompt():
  prompt_temp=PromptTemplate(template=prompt,input_variables=['context','question'])
  return prompt_temp
```

```python
def retrivel_answer(llm_model_name,prompt_name,v_db):
  qa=RetrievalQA.from_chain_type(
      llm=llm_model_name,
      chain_type='stuff',
      chain_type_kwargs={'prompt':prompt_name,'verbose':True},
      return_source_documents=True,
      retriever=v_db.as_retriever(search_kwargs={'k':2}))
  return qa
```

```python
def setup_model():
  embb=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2',model_kwargs={"device":"cpu"})
  vdb=FAISS.load_local("db/faiss",embb,allow_dangerous_deserialization= True)
  pmt=squ_prompt()
  get_answer=retrivel_answer(llm_model,pmt,vdb)
  return get_answer

```

```python
llm_qa_model=setup_model()
```
