```python
import requests
```

```python
#python client library code for fetching news
def fetch_newsapi(api_key, query, language='en', page_size=5):
    
    url = 'https://newsapi.org/v2/everything'
    params = {
        'q': query,                  
        'apiKey': api_key,          
        'language': language,
        'sort_by':'relevancy',
        'pageSize': page_size       
    }
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  
        data = response.json()
        #checking in response json from "everything" for status to be "ok"
        if data.get('status') != 'ok':
            print(f"Error: {data.get('message')}")
            return
        
        articles = data.get('articles', [])
        
        if not articles:
            print("No articles found.")
            return
        
        # Displaying the results
        for i, article in enumerate(articles):
            # numbering the article
            print(f"Article {i + 1}:")
            
            #fetching article author
            print(f"Author: {article.get('author')}")
            
            #fetching article title
            print(f"Title: {article.get('title')}")
            
            #fetching description      
            print(f"Description: {article.get('description')}")
            #get  URL
            print(f"URL: {article.get('url')}")
            
            #published date
            print(f"Publisted At:{article.get('publishedAt')}")
            
            print("-" * 60)
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

def main():
    #api key provided, generated from NewsAPI website
    api_key = '202cc8e87f0f44908e85483eecf40a44'   
    
    query = input("Enter the search query for news: ").strip()
    
    # Fetch and display news articles
    fetch_newsapi(api_key, query)

if __name__ == "__main__":
    main()
```
